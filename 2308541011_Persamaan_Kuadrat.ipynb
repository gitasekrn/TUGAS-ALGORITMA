```python
#1. Menentukan akar-akar persamaan kuadrat
import numpy as np
import matplotlib.pyplot as plt
print("Selamat datang! Pada program kali ini, kita akan menghitung akar-akar persamaan kuadrat")
print("jika diketahui persamaan ax**2+bx+c=0")
print("Langkah pertama yaitu memasukkan nilai a,b,c yang diketahui")
#Pengguna menginput nilai-nilai a,b,c dari persamaan yang diketahui
a_11 = float(input("Masukkan nilai a ="))
b_11 = float(input("Masukkan nilai b ="))
c_11 = float(input("Masukkan nilai c ="))
print("Persamaan yang diperoleh yaitu, {}x**2+{}x+{}".format(a_11,b_11,c_11))
print("==================================================================================================")
print("Langkah kedua yaitu menentukan determinan menggunakan rumus D=b-4ac")
#Membuat program untuk menghitung determinan
D_11 = b_11**2-(4*a_11*c_11)
print(f"Nilai determinan yang diperoleh yaitu D = {D_11}")
print("==================================================================================================")
print("Langkah terakhir yaitu menghitung akar-akar persamaan yang diperoleh, menggunakan rumus -b±√D/2a")
#Membuat program untuk menghitung akar-akar persamaan menggunakan cmath untuk menghitung akar
import cmath
x1_11 = (-b_11+cmath.sqrt(D_11))/2*a_11
x2_11 = (-b_11-cmath.sqrt(D_11))/2*a_11
print("Akar-akar persamaan yang diperoleh yaitu, ({:.0f},{:.0f})".format(x1_11,x2_11))
print("Sampai jumpa!")
```

```python
#2. Menentukan Determinan, apakah D=0, D<0, ataupun D>0
import numpy as np
import matplotlib.pyplot as plt
import cmath
print("Selamat datang! Kali ini kita akan menentukan apakah D=0, D<0, ataupun D>0")
print("Kita menentukan macam-macam determinan untuk mengetahui jenis akar-akar yang akan diperoleh")
print("Terdapat 3 jenis akar yang dapat diketahui melalui determinan, yaitu")
print("D=0 berarti kita memiliki akar-akar yang sama, D>0 berarti kita memiliki akar-akar berbeda(real), dan D<0 berarti kita memiliki akar-akar imajiner")
print("Pertama-tama masukkan nilai a,b,c yang diketahui")
a_11 = float(input("Masukkan nilai a ="))
b_11 = float(input("Masukkan nilai b ="))
c_11 = float(input("Masukkan nilai c ="))
print("Persamaan yang diperoleh yaitu, {}x**2+{}x+{}".format(a_11,b_11,c_11))
print("==================================================================================================")
print("Langkah kedua yaitu menentukan determinan menggunakan rumus D=b-4ac")
#Membuat program untuk menghitung determinan
D_11 = b_11**2-(4*a_11*c_11)
print(f'Determinan yang diperoleh adalah {D_11}')
print("==================================================================================================")
#Membuat tiga kemungkinan yang akan didapat
if D_11==0:
    print("Dari determinan yang diperoleh, kamu memperoleh akar-akar yang kembar")
    #Menghitung akar-akarnya
    x1_11 = (-b_11+cmath.sqrt(D_11))/2*a_11
    x2_11 = (-b_11-cmath.sqrt(D_11))/2*a_11
    print("Akar-akar yang diperoleh yaitu ({:.1f},{:.1f})".format(x1_11,x2_11))
elif D_11>0:
    print("Dari determinan yang diperoleh, kamu memperoleh akar-akar yang berbeda")
    #Menghitung akar-akarnya
    x1_11 = (-b_11+cmath.sqrt(D_11))/2*a_11
    x2_11 = (-b_11-cmath.sqrt(D_11))/2*a_11
    print("Akar-akar yang diperoleh yaitu ({:.1f},{:.1f})".format(x1_11,x2_11))
elif D_11<0:
    print("Dari determinan yang diperoleh, kamu memperoleh akar-akar imajiner")
    #Menghitung akar-akarnya
    x1_11 = (-b_11+cmath.sqrt(D_11))/2*a_11
    x2_11 = (-b_11-cmath.sqrt(D_11))/2*a_11
    print("Akar-akar yang diperoleh yaitu ({:.1f},{:.1f})".format(x1_11,x2_11))
print("Sampai jumpa!")
```

```python
#3. Membuat Grafik dari akar-akar persamaan kuadrat
import numpy as np
import matplotlib.pyplot as plt
import cmath

print("Selamat datang! Kali ini kita akan menentukan apakah D=0, D<0, ataupun D>0")
print("Kita menentukan macam-macam determinan untuk mengetahui jenis akar-akar yang akan diperoleh")
print("Terdapat 3 jenis akar yang dapat diketahui melalui determinan, yaitu")
print("D=0 berarti kita memiliki akar-akar yang sama, D>0 berarti kita memiliki akar-akar berbeda(real), dan D<0 berarti kita memiliki akar-akar imajiner")
print("Pertama-tama masukkan nilai a,b,c yang diketahui")
a_11 = float(input("Masukkan nilai a ="))
b_11 = float(input("Masukkan nilai b ="))
c_11 = float(input("Masukkan nilai c ="))
print("Persamaan yang diperoleh yaitu, {}x**2+{}x+{}".format(a_11,b_11,c_11))
print("==================================================================================================")
print("Langkah kedua yaitu menentukan determinan menggunakan rumus D=b-4ac")
#Membuat program untuk menghitung determinan
D_11 = b_11**2-(4*a_11*c_11)
print(f'Determinan yang diperoleh adalah {D_11}')
print("==================================================================================================")
#Membuat tiga kemungkinan yang akan didapat
if D_11==0:
    print("Dari determinan yang diperoleh, kamu memperoleh akar-akar yang kembar")
elif D_11>0:
    print("Dari determinan yang diperoleh, kamu memperoleh akar-akar yang berbeda")
elif D_11<0:
    print("Dari determinan yang diperoleh, kamu memperoleh akar-akar imajiner")

#Menghitung akar-akarnya
x1_11 = (-b_11+cmath.sqrt(D_11))/2*a_11
x2_11 = (-b_11-cmath.sqrt(D_11))/2*a_11
print("Akar-akar yang diperoleh dari determinan dan persamaan kuadrat yaitu ({:.3f},{:.3f})".format(x1_11.real,x2_11.real))

print("==================================================================================================")
print("Grafik yang diperoleh :")
#Lalu akan membuat program untuk membuat grafik
# 7. Format garis
x_11 = np.linspace(-20, 20, 500)
y_11 = a_11*x_11**2 + b_11*x_11 + c_11
plt.figure(figsize=(10, 8))
plt.plot(x_11, y_11, label=f'{a_11}x^2 + {b_11}x + {c_11} = 0')

if D_11>=0:
    plt.scatter([x1_11.real, x2_11.real], [0, 0], color='red')
    plt.text(x1_11.real, 0.5, f'({x1_11.real:.2f}, 0)', fontsize=12, ha='center')
    plt.text(x2_11.real, 0.5, f'({x2_11.real:.2f}, 0)', fontsize=12, ha='center')
else:
    plt.scatter([x1_11.real, x2_11.real], [x1_11.imag, x2_11.imag], color='red')
    plt.text(x1_11.real, x1_11.imag, f'({x1_11.real:.2f}, {x1_11.imag:.2f})', fontsize=12, ha='center')
    plt.text(x2_11.real, x2_11.imag, f'({x2_11.real:.2f}, {x2_11.imag:.2f})', fontsize=12, ha='center')

plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.title('Grafik Persamaan Kuadrat')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()

print("Sampai jumpa!")
```
