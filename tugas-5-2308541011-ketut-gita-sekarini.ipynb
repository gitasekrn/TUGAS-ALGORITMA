```python
print('Diberikan bilangan real 10 pertama, tentukan range!')
Z_11=range(10)
print(Z_11)
print('Lalu sebutkan elemen dari bilangan real < 10 tersebut!')
for elemen_11 in Z_11:
    print(elemen_11)
print('Bagaimana kalau bilangan real < 5?')
for elemen_11 in range (0,5):
    print(elemen_11)
print('Jabarkan bilangan real < 10 yang bernilai genap!')
for elemen_11 in range (0,10,2):
    print(elemen_11)
```
