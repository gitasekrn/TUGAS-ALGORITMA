```python
# Codingan untuk mencari gradien, persamaan garis, sekaligus grafik jika diketahui 4 titik
# namun pengguna dapat menginput titik-titiknya sendiri
import numpy as np
import matplotlib.pyplot as plt

# Pertama-tama pengguna dapat memasukkan titik A(g,i) dan B(t,a)
g_11 = input("Masukkan titik g =")
i_11 = input("Masukkan titik i =")
t_11 = input("Masukkan titik t =")
a_11 = input("Masukkan titik a =")

# Lalu untuk mencari gradien pada titik A dan B digunakan rumus Gradien1=a-i/t-g, maka
a_i_11= int(a_11)-int(i_11)
t_g_11= int(t_11)-int(g_11)
Gradien1_11=a_i_11/t_g_11
print("Gradien yang didapat pada garis pertama adalah {}".format(Gradien1_11))
print("Persamaan pada garis pertama adalah y={}x+{} ".format(Gradien1_11,float(-Gradien1_11)*float(g_11)+float(i_11)))
# Akan didapatkan persamaan garis pertama dengan rumus pada baris 16 diatas, y1 = Gradien1*(x-g)+i

# Selanjutnya pengguna akan memasukkan titik-titik C(s,e) dan D(r,n) terlebih dahulu
s_11 = input("Masukkan titik s =")
e_11 = input("Masukkan titik e =")
r_11 = input("Masukkan titik r =")
n_11 = input("Masukkan titik n =")

# Lalu untuk mencari gradien pada titik C dan D digunakan rumus yang sama yaitu Gradien2=n-e/r-s, maka
n_e_11= int(n_11)-int(e_11)
r_s_11= int(r_11)-int(s_11)
Gradien2_11=n_e_11/r_s_11
print("Gradien yang didapat pada garis kedua adalah {}".format(Gradien2_11))
print("Persamaan pada garis kedua adalah y={}x+{} ".format(Gradien2_11,float(-Gradien2_11)*float(s_11)+float(e_11)))
# Akan didapatkan persamaan garis kedua dengan rumus yang sama dengan baris 16, namun memakai variabel s dan e, yaitu y2 = Gradien2*(x-s)+e

# Selanjutnya akan ditentukan titik potong, maka Gradien1 tidak boleh sama dengan Gradien2, karena jika kedua persamaan garis memiliki gradien yang sama
# Maka garis tidak akan memiliki titik potong, kedua garis akan sejajar, jadi
if Gradien1_11==Gradien2_11:
    print("Kedua persamaan garis memiliki gradien yang sama, hal tersebut menyebabkan kedua garis tidak memiliki titik potong atau sejajar.")
else:
    # Jika Gradien pada kedua persamaan garis memiliki jumlah yang berbeda, maka akan dicari titik potongnya dengan = (TitikpotongX,TitikpotongY)
    titikpotongX_11 = (float(Gradien1_11)*float(g_11)-float(i_11)-float(Gradien2_11)*float(s_11)+float(e_11))/(float(Gradien1_11)-float(Gradien2_11))
    titikpotongY_11 = float(Gradien2_11)*(titikpotongX_11-float(s_11))+float(e_11)
# Akan didapat titik potong berupa (x,y)

print("Titik potong yang didapat dari persamaan garis diatas adalah ({},{})".format(titikpotongX_11,titikpotongY_11))
    
# Lalu akan dibuat code untuk membuat dan meletakkan garis pada grafiknya
x=np.arange(titikpotongX_11-5,titikpotongX_11+5,0.01)
plt.figure(figsize=(4,4),dpi=500)
plt.title('Grafik persamaan garis berpotong')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.xlabel('X')
plt.ylabel('Y')
plt.plot(x,float(Gradien1_11)*(x-float(g_11))+float(i_11),color='blue',label='y={}x+{}'.format(Gradien1_11,float(-Gradien1_11)*float(g_11)+float(i_11)))
plt.plot(x,float(Gradien2_11)*(x-float(s_11))+float(e_11),color='red',label='y={}x+{}'.format(Gradien2_11,float(-Gradien2_11)*float(s_11)+float(e_11)))
plt.plot(titikpotongX_11,titikpotongY_11,color='purple',marker='o',markersize=4)
plt.legend(loc='best')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

print("Mari kita mencari gradien dan persamaan garis jika diketahui 2 titik")
print("Jika diketahui Y(0,3) dan Z(2,11)")

#Didefinisikan titik Y(0,3) dan Z(2,11)
a_11=0 #Titik Y(a,b)
b_11=3
c_11=2 #Titik Z(c,d)
d_11=11

#Menentukan gradien
Gradien1_11=(d_11-b_11)/(c_11-a_11)
print(f'Gradien yang didapat dari 2 titik tersebut adalah {Gradien1_11}')
print("Persamaan garis kedua adalah y={}x+{} ".format(Gradien1_11,-Gradien1_11*a_11+b_11))

#Menentukan rentang x
x_11=np.linspace(-7,7,300) 

#Membuat grafik dan menentukan warna, membuat label
plt.plot(x_11,Gradien1_11*(x_11-a_11)+b_11, color='red', label='y={}x+{}'.format(Gradien1_11,-Gradien1_11*a_11+b_11))

#Membuat judul grafik
plt.title('Grafik dari satu persamaan garis')

#Membuat garis vertikal dan horizontal pada grafik
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)

#Memunculkan grid pada grafik
plt.grid(True)

plt.xlabel('X')
plt.ylabel('Y')

#Memunculkan hasil
plt.show()
```
