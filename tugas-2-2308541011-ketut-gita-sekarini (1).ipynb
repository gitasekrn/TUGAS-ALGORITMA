```python
##
# 2.9
print("learning about 'and', 'or', 'not' in phyton")
print("Funfact 1")
a = input("put a random number for 'a'")
b = input("put a random number for 'b'")
a_b = (int(a)<int(b) and int(a)>int(b))
print(f"{a}<{b} and {a}>{b} are")
print(a_b)
print("Do you know why its false?")
print("Because in phyton 'and' will say its true if both of them are correct")

print("Funfact 2")
c = input("put a random number for 'c'")
d = input("put a random number for 'd'")
c_d = (int(c)<int(d) or int(c)>int(d))
print(f"{c}<{d} or {c}>{d} are")
print(c_d)
print("So 'and' != 'or', because 'or' will be correct even though one of them is false. 'or' only need one of them to be correct.")

print("Funfact 3")
e = input("put a random number for 'e'")
f = input("put a random number for 'f'")
e_f = (not (int(e)<int(f) or int(e)>int(f)))
print(f" 'not' ({e}<{f} or {e}>{f} are)")
print(e_f)
print("So, this one is a bit different from the other two, 'not' in phyton will tell you the opposite of the right answer. 'True' will be 'False', and the other way around")
```
