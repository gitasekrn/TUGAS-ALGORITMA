```python
#Menentukan nilai n yang memenuhi menggunakan akar positif(x1)
import cmath
print("Diketahui x memenuhi")
print("x**2+x-1=0")
print("Tentukan nilai yang memenuhi")
print("x**2024 - x**2025 = (x(1-x))**n")
print("Diketahui a=1;b=1;c=-1")
print("Akan dicari nilai x1 dengan rumus:")
print("-1 + √b**2 - 4*a*c / 2*a")
print("=====================================================================================")
#Mencari nilai x1 (akar positif)
x1_11 = (-1 + cmath.sqrt(5)) / 2
print(f'Nilai x yang diperoleh adalah {x1_11.real}')
#Mencari nilai n yang memenuhi jika akar positif(x1)
#Mencari menggunakan logaritma(log)
#log(x**2024-x**2025) dapat diubah kedalam bentuk 2024*log(x)+log(1-x), sehingga
logkiri1_11 = (2024*(cmath.log(x1_11))) + cmath.log(1-(x1_11))
#Lalu log(x(1-x)) dapat diubah kedalam bentuk log(x)+log(1-x), maka
logkanan1_11 = cmath.log(x1_11) + cmath.log(1-(x1_11))
#Lalu dapat ditemukan nilai n yang memenuhi persamaan diatas
n1_11 = logkiri1_11/logkanan1_11
print(f'Nilai n yang memenuhi akar positif adalah {n_11.real}')
```

```python
#Menentukan nilai n yang memenuhi menggunakan akar negatif(x2)
import cmath
print("Diketahui x memenuhi")
print("x**2+x-1=0")
print("Tentukan nilai yang memenuhi")
print("x**2024 - x**2025 = (x(1-x))**n")
print("Diketahui a=1;b=1;c=-1")
print("Akan dicari nilai x1 dengan rumus:")
print("-1 - √b**2 - 4*a*c / 2*a")
print("=====================================================================================")
#Mencari nilai x1 (akar positif)
x2_11 = (-1 - cmath.sqrt(5)) / 2
print(f'Nilai x yang diperoleh adalah {x2_11.real}')
#Mencari nilai n yang memenuhi jika akar positif(x1)
#Mencari menggunakan logaritma(log)
#log(x**2024-x**2025) dapat diubah kedalam bentuk 2024*log(x)+log(1-x), sehingga
logkiri2_11 = (2024*(cmath.log(abs(x2_11)))) + cmath.log(abs(1-(x2_11)))
#Lalu log(x(1-x)) dapat diubah kedalam bentuk log(x)+log(1-x), maka
logkanan2_11 = cmath.log(abs(x2_11)) + cmath.log(abs(1-(x2_11)))
#Lalu dapat ditemukan nilai n yang memenuhi persamaan diatas
n2_11 = logkiri2_11/logkanan2_11
print(f'Nilai n yang memenuhi n adalah {n2_11.real}')
```
