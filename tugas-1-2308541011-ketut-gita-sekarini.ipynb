```python
##
# Celcius ke Farenheit converter
celcius = float(input("Berapa derajat celcius hari ini? = "))
farenheit = celcius * 1.8 + 32

print(f"suhu {celcius} C sama dengan {farenheit} F")
if farenheit >= 95 :
    print ("Cuaca hari ini terdeteksi panas. Jangan lupa meminum air yang cukup!")
    
if farenheit < 95 and farenheit >= 69 :
    print ("Cuaca hari ini terdeteksi normal. Jangan lupa keluar rumah dan menikmati hari!")
    
if farenheit <= 68 :
    print ("Cuaca hari ini terdeteksi dingin. Jangan lupa membawa jaket yang tebal!")
```
